```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import json

# Import API key
from api_keys import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
locations_df = pd.read_csv("weather.csv")
#print(locations_df.info())

# When reading in the dataframe I found that the read function converts anything that is NA in the country
# column to NaN. This is incorrect since NA is the country code for Namibia
w1 = locations_df[locations_df.isna().any(axis=1)]
#print(w1)

# Change the NaN in the country column to NA
locations_df["Country"] = locations_df["Country"].fillna("NA")
locations_df[locations_df["Country"] == "NA"]
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Storing Lat and Long values
location = locations_df[["Lat", "Lon"]]

print(type(location))

# Store humidity series
humid = locations_df["Humidity"].astype(float)
```

```python
# Plot heatmap

#Create heat layer
heat_layer1 = gmaps.heatmap_layer(location, weights=humid, dissipating=False, max_intensity=100,point_radius=1)

figure_layout = {'width': '300px','height': '1000px','border': '0px solid black','padding': '1px'}

fig1 = gmaps.figure(figure_layout)

# Add layer
fig1.add_layer(heat_layer1)

# Display figure
fig1
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
location_refined = locations_df[(locations_df["Max Temp"] < 80) & (locations_df["Max Temp"] > 70)
                               & (locations_df["Wind Speed"] < 10) & (locations_df["Cloudiness"] == 0)]
location_refined
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = pd.DataFrame()
hotel_df["Hotel Name"] = ""
hotel_df["City"] = ""
hotel_df["Country"] = ""
hotel_df["Lat"] = ""
hotel_df["Lng"] = ""
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

target_radius = 5000
target_type = "lodging"

for index, row in location_refined.iterrows():
    lat = row["Lat"]
    lng = row["Lon"]
    lct = str(lat)+", "+str(lng)
    params = {
        "location": lct,
        "radius": target_radius,
        "type": target_type,
        "key": g_key,
    }
    resp = requests.get(base_url, params=params).json()
    #print(json.dumps(resp, indent=4, sort_keys=True))
    #print(lct)
    try:
        #print(resp["results"][1]["name"])
        print(json.dumps(resp["results"][1], indent=4, sort_keys=True))
        hotel_df.loc[index, "Hotel Name"] = resp["results"][1]["name"]
        hotel_df.loc[index, "Vicinity"] = resp["results"][1]["vicinity"]
        hotel_df.loc[index, "Lat"] = round(resp["results"][1]["geometry"]["location"]["lat"], 2)
        hotel_df.loc[index, "Lng"] = round(resp["results"][1]["geometry"]["location"]["lng"], 2)
    except (KeyError, IndexError):
        print("Missing field/result....skipping.")
    print("---------------")
```

```python
# Testing output of hotel_df
hotel_df
```

```python
# Add City and Country info from locations dataframe
for ids in hotel_df.index.values.tolist():
    hotel_df.loc[ids, ["City", "Country"]] = location_refined.loc[ids, ["City", "Country"]]
    
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig1.add_layer(markers)
# Display Map
fig1
```
